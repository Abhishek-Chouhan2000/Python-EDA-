# using IQR Technique 

```python
# Writing a outlier function for removing outliers in important Columns
def iqr_technique(DFcolumn):
    Q1 = np.percentile(DFcolumn , 25)
    Q3 = np.percentile(DFcolumn , 75)
    IQR = Q3 - Q1 
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)

    return lower_range , upper_range 
```

```python
lower_bound , upper_bound = iqr_technique(porter_df['price_rs'])
porter_df = porter_df[(porter_df.price_rs >lower_bound) & (porter_df.price_rs < upper_bound)]
```

# ***6. Does delivery time follow normal distribution (Porter)?***

```python
from scipy.stats import norm
porter_time = porter_df[porter_df['competitor_company'] =='Porter']['delivery_time_min']
mu, sigma = norm.fit(porter_time)

plt.figure(figsize=(8,5))
sns.histplot(porter_time,bins=40 , stat = 'density')
x = np.linspace(porter_time.min(), porter_time.max(),100)
plt.plot(x,norm.pdf(x,mu,sigma),'r')
plt.title("Porter Delivery Time Distribution (Gaussian Fit)")
plt.show()
```

# **Observation**
**Porterâ€™s delivery time distribution closely approximates a normal distribution, validating the use of Gaussian modeling. Minor skewness is observed due to delayed and failed deliveries. This confirms that most deliveries cluster around the mean with predictable variability. Statistical modeling is therefore appropriate for performance evaluation.**

# ***7 & 8. Î¼ and Ïƒ comparison across companies***

### **CREATE stats_df dateframe

```python

stats = []

for competitor_company in porter_df['competitor_company'].unique():
    data = porter_df[porter_df['competitor_company'] == competitor_company]['delivery_time_min']
    mu, sigma = norm.fit(data)
    
    stats.append({
        'competitor_company': competitor_company,
        'mean': mu,
        'std': sigma
    })

stats_df = pd.DataFrame(stats)
stats_df
```

```python
stats_df.columns
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=stats_df,
    x='mean',
    y='std',
    hue='competitor_company',
    s=120
)

plt.title("Mean vs Standard Deviation of Delivery Time")
plt.xlabel("Mean Delivery Time (minutes)")
plt.ylabel("Standard Deviation (Consistency)")
plt.grid(True)
plt.show()

```

# **Observation**

* Companies on the left are faster (lower mean delivery time)

* Companies at the bottom are more consistent (lower std deviation)

* Blue Dart typically appears lower on the Y-axis â†’ highest reliability

* Porter sits in the middle â†’ balanced speed and consistency

* Shadowfax often has low mean but high std â†’ fast but inconsistent

# ***16.Best price-to-performance ratio***

### **KPI Creation**

```python
kpi = porter_df.groupby('competitor_company').agg(
    avg_time=('delivery_time_min', 'mean'),
    avg_price=('price_rs', 'mean'),
    avg_rating=('customer_rating', 'mean'),
    on_time_rate=('status', lambda x: (x == 'delivered').mean() * 100)
).reset_index()

kpi
```

```python
kpi['price_to_time'] = kpi['avg_price'] / kpi['avg_time']

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=kpi,
    x='avg_price',
    y='avg_time',
    size='price_to_time',
    hue=kpi.index,
    sizes=(100,800)
)
plt.title("Price vs Delivery Time (Bubble = Value)")
plt.show()
```

# **Observation** 

**Porter offers the best price-to-performance ratio among competitors. It delivers competitive speed at moderate pricing, outperforming both premium and low-cost players on value metrics. Blue Dart excels in reliability but at a higher cost. Porterâ€™s balance makes it attractive for SMEs and frequent users.**
